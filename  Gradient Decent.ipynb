# Linear Regression using Gradient Decent

```python
m = 0
c = 0
lr = .0001 #learning rate
n = len(df.x)

for i in range(500):
    y_hat = (m*x)+c
    Dm = (-2/n)*sum(x*(y-y_hat))
    Dc = (-2/n)*sum(y-y_hat)

    #update the weight
    m = m - (lr*Dm)
    c = c - (lr*Dc)
    print(m,c)

print(m,c)
```
